# Exploratory Data Analysis

```python
# We can extract a new feature from The Location which is the country where the study hold
countries = [ str(data.Locations.iloc[i]).split(',')[-1] for i in range(data.shape[0])]
data['Country'] = countries
```

```python
data.columns
```

```python
data['Country'].value_counts()
```

```python
# Impute Interventions , Phases , Locations by Missing Category
categorical_features = data.select_dtypes(include =object).columns

features = categorical_features[data[categorical_features].isnull().mean() > 0]

for feature in features:
   data[feature] = data[feature].fillna(f"Missing {feature}")
```

```python
# Detecting (Percentage) Missing Data

data.isnull().mean() * 100
```

```python
# Now the Time to handle The missing data for the Enrollment
 
# Check the skewness in Enrollment column

data.Enrollment.skew()
```

```python
# Plotting the distribution of the enrollment
data.Enrollment.plot(kind = 'kde')
```

```python
# Now we will use the median value to impute the missing values
median_value = data.Enrollment.median()
print(median_value)
```

```python
#Use the Median to impute Missing Values
data.Enrollment = data.Enrollment.fillna(median_value)
```

```python
# Detecting (Percentage) Missing Data

data.isnull().mean() * 100
```

```python
plt.figure(figsize=(8,5))
plt.bar(top_10_Countries.index, top_10_Countries.values)
# plt.xlabel('Countries')
plt.ylabel('Contribution')
plt.title('Top 10 Countries')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```
